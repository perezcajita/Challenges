```python
import pandas as pd

df_pd = pd.read_csv("/Workspace/Users/javier.perez5739@alumnos.udg.mx/Challenge intermedio/winequality-red.csv", sep=";")
df_red = spark.createDataFrame(df_pd)
df_red.display()
```

```python
# Creamos una nueva columna llamada label que serÃ¡ nuestra variable objetivo para clasificaciÃ³n binaria. Consideramos que un vino es de buena calidad si su calificaciÃ³n 

from pyspark.sql.functions import col

df_red = df_red.withColumn("label", (col("quality") >= 6).cast("int"))
df_red.select("quality", "label").groupBy("label").count().show()
```

```python
# Usamos VectorAssembler para combinar todas las columnas numÃ©ricas (excepto quality y label) en una sola columna de vectores llamada features, que es el formato que Spark ML requiere para entrenar modelos.

from pyspark.ml.feature import VectorAssembler

# Seleccionar las columnas que serÃ¡n usadas como variables predictoras
features = [c for c in df_red.columns if c not in ("quality", "label")]

# Ensamblar las columnas en un solo vector
assembler = VectorAssembler(inputCols=features, outputCol="features")
df_model = assembler.transform(df_red).select("features", "label")
```

```python
# Dividimos los datos en dos subconjuntos: uno para entrenar el modelo y otro para evaluarlo.

train, test = df_model.randomSplit([0.8, 0.2], seed=42)
```

```python
# Entrenamos un modelo de clasificaciÃ³n binaria usando regresiÃ³n logÃ­stica, una tÃ©cnica comÃºn para problemas de clasificaciÃ³n

from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(featuresCol="features", labelCol="label")
model = lr.fit(train)

# El modelo aprende a predecir si un vino es de buena calidad (1) o no (0) a partir de sus caracterÃ­sticas fisicoquÃ­micas
```

```python
# Evaluamos el modelo usando el conjunto de prueba y calculamos una mÃ©trica de exactitud basada en el Ã¡rea bajo la curva ROC (AUC).

from pyspark.ml.evaluation import BinaryClassificationEvaluator

predictions = model.transform(test)
evaluator = BinaryClassificationEvaluator(labelCol="label")
accuracy = evaluator.evaluate(predictions)

print(f"Exactitud del modelo: {accuracy:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir el DataFrame de Spark a Pandas para graficar
df_plot = df_red.select("label").groupBy("label").count().toPandas()

# Renombrar valores para claridad
df_plot["label"] = df_plot["label"].map({0: "Mala calidad", 1: "Buena calidad"})

# Crear grÃ¡fica de barras
plt.figure(figsize=(6,4))
plt.bar(df_plot["label"], df_plot["count"], color=["tomato", "forestgreen"])
plt.title("DistribuciÃ³n de calidad del vino (red)")
plt.xlabel("CategorÃ­a")
plt.ylabel("Cantidad de muestras")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#  Visualizar en 3D los datos de vino tinto y blanco usando tres variables (por ejemplo: alcohol, volatile acidity, density) y mostrar un plano de decisiÃ³n aproximado generado por un modelo de regresiÃ³n logÃ­stica.
```

```python
from pyspark.sql.functions import col

# Para vino tinto: buena calidad si quality >= 6
df_red = df_red.withColumn("label", (col("quality") >= 6).cast("int"))

# Para vino blanco: buena calidad si quality >= 7
df_white = df_white.withColumn("label", (col("quality") >= 7).cast("int"))
```

```python
df_all = df_red.unionByName(df_white)
```

```python
# Seleccionar tres variables para grafica

cols_3d = ["alcohol", "volatile acidity", "density", "label"]
df_3d = df_all.select(*cols_3d).toPandas()
```

```python
# Entrenar el modelo de regresiÃ³n, 

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

X = df_3d[["alcohol", "volatile acidity", "density"]]
y = df_3d["label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = LogisticRegression()
model.fit(X_scaled, y)
```

```python
# GrÃ¡fica interactiva en 3D con un hiperplano, 

import plotly.graph_objects as go
import numpy as np

# Crear malla para el hiperplano
xx, yy = np.meshgrid(np.linspace(-2, 2, 30), np.linspace(-2, 2, 30))
coef = model.coef_[0]
intercept = model.intercept_[0]
zz = (-coef[0]*xx - coef[1]*yy - intercept)/coef[2]

# Crear figura
fig = go.Figure()

# Agregar puntos de datos
fig.add_trace(go.Scatter3d(
    x=X_scaled[:,0],
    y=X_scaled[:,1],
    z=X_scaled[:,2],
    mode='markers',
    marker=dict(
        size=4,
        color=y,
        colorscale='RdYlGn',
        opacity=0.7
    ),
    name='Vinos'
))

# Agregar hiperplano
fig.add_trace(go.Surface(
    x=xx,
    y=yy,
    z=zz,
    showscale=False,
    opacity=0.4,
    colorscale='Blues',
    name='Hiperplano'
))

# Etiquetas
fig.update_layout(
    scene=dict(
        xaxis_title='Alcohol',
        yaxis_title='Volatile Acidity',
        zaxis_title='Density'
    ),
    title='ClasificaciÃ³n de vinos con hiperplano de regresiÃ³n logÃ­stica',
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```
